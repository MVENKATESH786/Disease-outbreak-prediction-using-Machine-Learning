```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
data_path = r"C:\Users\Venky\Desktop\Disease-outbreak-prediction-using-Machine-Learning\Datasets\parkinsons.csv"
data = pd.read_csv(data_path)

# Separate features (X) and target (Y)
X = data.drop(columns=['status'])  # Assuming 'status' is the target column
Y = data['status']

# If you have applied resampling, make sure these variables exist
# Otherwise, use X and Y directly
X_resampled, Y_resampled = X, Y  # Replace this with actual resampling if done

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(
    X_resampled, Y_resampled, test_size=0.2, stratify=Y_resampled, random_state=2
)

# Verify the shape of the splits
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
print(data.isnull().sum())
```

```python
# Display the shapes of the splits
print("\nShapes of the data splits:")
print(f"X: {X.shape}")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
```

```python
print(X_train.dtypes)
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert categorical columns to numerical
for col in X_train.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    # Apply same transformation to test data
    X_test[col] = le.transform(X_test[col])
```

```python
model = svm.SVC(random_state=2)
model.fit(X_train, Y_train)
```

```python
# Train the model
best_model = svm.SVC(random_state=2)
best_model.fit(X_train, Y_train)

# Calculate training accuracy
train_accuracy = best_model.score(X_train, Y_train)
print(f"\nTraining Accuracy: {train_accuracy * 100:.2f}%")
```

```python
print(X_test.dtypes)
```

```python
# Ensure all features are numeric
X_test = X_test.select_dtypes(include=[np.number])

# Make predictions
Y_pred = best_model.predict(X_test)

# Calculate test accuracy
test_accuracy = accuracy_score(Y_test, Y_pred)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
# Classification report
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred))

# Confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(Y_test, Y_pred))
```

```python
# Build a predictive system
sample_data = np.array([X_test[0]])  
sample_prediction = best_model.predict(sample_data)
print("\nSample Prediction:", "Parkinson's" if sample_prediction[0] == 1 else "No Parkinson's")
```

```python
# Save the model to a .sav file
model_filename = "parkinsons_model.sav"
with open(model_filename, 'wb') as model_file:
    pickle.dump(best_model, model_file)

print(f"\nModel saved to {model_filename}")
```
